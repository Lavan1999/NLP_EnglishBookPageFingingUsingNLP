```python
import pandas as pd

# Initialize an empty list to store extracted text
extracted_text = []

for i, page in enumerate(read.pages):
    # Extract text from the current page
    content = page.extract_text()
    # Apply cleaning to the extracted text
    content = content.replace("\n", "").replace("\t", "").replace("   ", "")
    # Append the cleaned text along with the corresponding index value to the list
    extracted_text.append({'x_feature': content, 'y_target': i})

df = pd.DataFrame(extracted_text)

print(df)
```

# **NLP-Using Sklearn**

```python
from sklearn.feature_extraction.text import CountVectorizer #this ll convert text into vector form matrix(token counts)
```

```python
b=CountVectorizer(ngram_range=(1,1))
train_x = b.fit_transform(df['x_feature'])
train_x.toarray()
```

# **Machine Learning**

```python
from sklearn.svm import SVC
```

```python
y = df['y_target']
```

```python
model=SVC().fit(train_x,y)
```

```python
test_x = b.transform(["hi"])
model.predict(test_x)
```
